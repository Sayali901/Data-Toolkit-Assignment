2)How does broadcasting work in NumPy?
  - Broadcasting simplifies mathematical operations on arrays with different shapes. It enables NumPy to efficiently apply operations element-wise without explicitly copying or reshaping data.

  It automatically adjusts the smaller array to match the shape of the larger array by replicating its values along the necessary dimensions. This feature reduces memory usage and eliminates the need for manual loops making code concise and computationally faster making it essential for handling large datasets and performing complex calculations in python.

8)What does the term “vectorized operation” mean in NumPy?

Vectorization is a key principle for optimizing numerical computations in NumPy and is crucial for performance in scientific, computing,data analysis and machine learning.

18)What are some advanced features of NumPy?
   - 1. Broadcasting:
Broadcasting allows NumPy to perform operations on arrays of different shapes by automatically resizing smaller arrays to match the shape of the larger array.
This avoids the need for explicit array reshaping and improves code efficiency.
It's particularly useful when dealing with arrays that have a different number of dimensions or when performing operations on arrays with different shapes.
2. Universal Functions (ufuncs):
Ufuncs are functions that operate on each element of a NumPy array in an element-wise manner.
They offer a concise and efficient way to perform mathematical, logical, and other operations on arrays.
Examples include np.add(), np.subtract(), np.sqrt(), and np.sin().
3. Vectorization:
Vectorization leverages NumPy's optimized functions to perform operations on entire arrays at once, rather than relying on slow Python loops.
This significantly speeds up computations, especially for large arrays.
It's a key principle in NumPy for writing efficient and readable code.
4. Advanced Array Manipulation:
NumPy provides a rich set of functions for manipulating arrays, including reshaping, slicing, indexing, and combining arrays.
Reshaping: Changing the shape of an array (e.g., from a 1D to a 2D array).
Slicing: Extracting subsets of an array.
Indexing: Accessing specific elements or groups of elements within an array.
Concatenating and Stacking: Combining multiple arrays into a single larger array.
5. Linear Algebra:
NumPy provides extensive support for linear algebra operations, including matrix operations, solving linear equations, and more.
This is essential for various scientific and engineering applications.
6. Advanced Indexing:
NumPy supports advanced indexing techniques, such as boolean indexing, which allows for accessing elements based on conditions.
This provides flexible and powerful ways to select subsets of data based on specific criteria.
7. Other Advanced Features:
Sorting and Searching:
NumPy offers functions for sorting and searching arrays, including functions like np.sort() and np.searchsorted().

```python
#How do you create a 2D NumPy array and calculate the sum of each row
import numpy as np
arr = np.array([[1,2,3],[4,5,6]])
def row_sum(row):
  return np.sum(row)
result = np.apply_along_axis(row_sum,axis=1,arr=arr)
print(result)
```

```python
#How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create a sample DataFrame
df = pd.DataFrame({
    'Math': [90, 85, 80, 75, 70],
    'Science': [85, 80, 75, 70, 65],
    'English': [80, 75, 70, 65, 60]
})

# Compute the correlation matrix
corr_matrix = df.corr()

# Create a heatmap to visualize the correlation matrix
sns.heatmap(corr_matrix, annot=True)

# Add a title
plt.title('Subject Correlation Heatmap')

# Display the plot
plt.show()
```
